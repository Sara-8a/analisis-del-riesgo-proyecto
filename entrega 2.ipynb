# **Explicación**

---

## 1. Datos, limpieza y retornos
Sea $\mathcal T=\{1,\dots,T\}$ el índice temporal (días hábiles) y $\mathcal J=\{1,\dots,N\}$ el conjunto de activos. Con precios ajustados $P_t^{(j)}$, los **retornos logarítmicos** diarios son
$$
 r_t^{(j)} = \ln\!\Big(\tfrac{P_t^{(j)}}{P_{t-1}^{(j)}}\Big), \qquad t\in\mathcal T,\; j\in\mathcal J.
$$
Propiedades: (i) *aditividad* $\sum_{t=1}^T r_t^{(j)} = \ln P_T^{(j)} - \ln P_0^{(j)}$; (ii) para retornos simples $R_t^{(j)}=e^{r_t^{(j)}}-1\approx r_t^{(j)}$ si $|r_t^{(j)}|\ll 1$; (iii) coherencia con composición geométrica.

**Anualización.** Con $D=252$ días hábiles,
$$
 \mathbb E\big[R^{\mathrm{ann}}\big] \approx e^{D\,\bar r}-1, \qquad \sigma^{\mathrm{ann}} = \sqrt{D}\, s_r.
$$
Por Jensen, $\mathbb E[e^X]-1 \ge e^{\mathbb E[X]}-1$; se usa $e^{D\bar r}-1$ como estimador *plug-in* para interpretación económica y SML.

---

## 2. Portafolio y riqueza
Sea $  w=(w_1,\dots,w_N)^\top$ con $\sum_j w_j=1$. Bajo rebalanceo a pesos objetivo, el retorno log del portafolio es
$$
 r_t^{(p)} = \sum_{j=1}^N w_j\, r_t^{(j)}.
$$
La **riqueza relativa** (normalizada a $1$ en $t=0$) es
$$
 W_t^{(p)} = \exp\!\Big(\sum_{s\le t} r_s^{(p)}\Big), \qquad \mathrm{DD}_t = \frac{W_t^{(p)}}{\max_{s\le t} W_s^{(p)}}-1, \qquad \mathrm{MDD}=\min_t \mathrm{DD}_t.
$$

**Beta del portafolio.** Con retorno de mercado $r_t^{(m)}$ y $\beta_i=\mathrm{Cov}(r_i,r_m)/\mathrm{Var}(r_m)$, si $w$ es fijo:
$$
 \beta_p = \frac{\mathrm{Cov}(r^{(p)},r^{(m)})}{\mathrm{Var}(r^{(m)})} = \sum_{j=1}^N w_j\,\beta_j.
$$

---

## 3. Métricas de riesgo–retorno
Para cada serie $r_t$:
$$\begin{aligned}
 R &:= e^{D\,\bar r}-1, & \sigma &:= \sqrt{D}\, s_r, & \text{Sharpe} &:= \frac{R-R_f}{\sigma},\\
 \text{Sortino} &:= \frac{R-R_f}{\sigma_-}, & \text{Calmar} &:= \Big|\tfrac{R-R_f}{\mathrm{MDD}}\Big|, & \text{Treynor} &:= \tfrac{R-R_f}{\beta},\\
 \mathrm{TE} &:= \sqrt{D}\,\mathrm{sd}(r_p-r_m), & \mathrm{IR} &:= \tfrac{\alpha}{\mathrm{TE}}. &&
\end{aligned}$$

**Diagnósticos:** **Jarque–Bera** (asimetría y curtosis) y **Ljung–Box** (autocorrelación, lag 10) sobre $r_t$.

---

## 4. CAPM, alfa de Jensen y SML con intercepto fijo
**CAPM poblacional:**
$$
 \mathbb E[R_i] = R_f + \beta_i\big(\mathbb E[R_m]-R_f\big), \qquad \beta_i = \frac{\mathrm{Cov}(r_i,r_m)}{\mathrm{Var}(r_m)}.
$$
**Alfa de Jensen:** con $\hat R_\cdot$ vía $e^{D\bar r}-1$,
$$
 \hat\alpha_i = \hat R_i - \Big(R_f + \hat\beta_i\big(\hat R_m - R_f\big)\Big).
$$
**SML teórica ($R_f=0.03$):**
$$
 y(\beta) = R_f + (\hat R_m - R_f)\, \beta.
$$
**Regresión anclada:** dados $(\beta_i,y_i)$ con $y_i=\hat R_i$,
$$
 y_i = R_f + b\,\beta_i + \varepsilon_i \;\Longleftrightarrow\; z_i := y_i-R_f = b\,\beta_i+\varepsilon_i,\quad \hat b = \frac{\sum_i \beta_i z_i}{\sum_i \beta_i^2},\quad \hat y_i= R_f + \hat b\,\beta_i.
$$
$$ R^2 = 1 - \frac{\sum_i (y_i-\hat y_i)^2}{\sum_i (y_i-\bar y)^2}. $$
Interpretación: si $\hat b < (\hat R_m - R_f)$, el panel empírico presenta prima sistemática menor que la teórica.

---

## 5. CUSUM de residuos OLS y eventos extremos
Modelo con constante $r_t=\mu+\varepsilon_t$. Con residuos recursivos $w_t$ y $\hat\sigma_w$,
$$
 W_t = \sum_{s=k+1}^t \frac{w_s}{\hat\sigma_w}, \qquad t=k+1,\dots,T,\; k=1.
$$
Cruces de bandas críticas sugieren ruptura. **Extremos**: marcar $|z_t|\ge 3$ con $z_t=(r_t-\bar r)/\hat\sigma$.

---

## 6. BOS (Break of Structure) vía *swings*
Para una serie $S_t$ (precio o riqueza) y ventana $q$:
$$\begin{aligned}
 S_t \text{ es swing-high } &\iff S_t > S_{t-i}\ \&\ S_t > S_{t+i}\quad \forall i=1,\dots,q,\\
 S_t \text{ es swing-low } &\iff S_t < S_{t-i}\ \&\ S_t < S_{t+i}\quad \forall i=1,\dots,q.
\end{aligned}$$
Con niveles últimos $H_{t-1},L_{t-1}$ (\*sin mirar la vela actual*),
$$ \mathrm{BOS\_up}:\ S_t>H_{t-1}, \qquad \mathrm{BOS\_down}:\ S_t<L_{t-1}. $$

---

## 7. Consideraciones de estimación y robustez
- **Log→aritmético (sesgo de Jensen):** si $r_t\sim\mathcal N(\mu,\sigma^2)$, entonces $\mathbb E[R^{\mathrm{ann}}]=e^{D\mu+\tfrac{1}{2}D\sigma^2}-1$.
- **HAC (Newey–West):** para errores estándar robustos en $\beta,\alpha$ bajo $\mathrm{AC}/\mathrm{HC}$.
- **Rolling:** $\beta_t,\alpha_t$ en ventanas móviles para detectar cambios de régimen.

---

## 8. Complejidad y riesgos de datos
- **Complejidad:** descarga/limpieza $\mathcal O(TN)$; covarianzas $\mathcal O(TN)$; BOS $\mathcal O(T)$ por serie; CUSUM $\mathcal O(T)$.
- **Datos:** precios ajustados (dividendos/splits). Posible sesgo de supervivencia. No se modelan costos ni restricciones de margen.

---

## Apéndice A. Derivaciones
**A.1 Regresión anclada.** Minimizar $\sum_i (z_i-b\beta_i)^2$ ⇒ $\partial/\partial b = -2\sum_i \beta_i(z_i-b\beta_i)=0\Rightarrow$ fórmula de $\hat b$ en la fórmula mostrada arriba.

**A.2 Beta del portafolio.** De la fórmula mostrada arriba y linealidad de covarianza: $\mathrm{Cov}(\sum_j w_j r_j, r_m) = \sum_j w_j\,\mathrm{Cov}(r_j,r_m)$; dividir por $\mathrm{Var}(r_m)$ da la fórmula mostrada arriba.

**A.3 Rebalanceo vs. buy-and-hold.** Si no se rebalancea, en general $r_t^{(p)}\neq\sum_j w_j r_t^{(j)}$; la identidad la fórmula mostrada arriba aproxima re-balanceo frecuente.


```python
import numpy as np
import pandas as pd
import yfinance as yf
import matplotlib.pyplot as plt
from scipy import stats
from scipy.signal import argrelextrema  # para BOS
import statsmodels.api as sm
from statsmodels.stats.diagnostic import breaks_cusumolsresid, acorr_ljungbox
from typing import List, Dict, Optional, Tuple, Any
import warnings
import logging
import time

warnings.filterwarnings('ignore')  # silenciar warnings no críticos
logging.basicConfig(level=logging.ERROR)
logger = logging.getLogger(__name__)

try:
    from IPython.display import display
except ImportError:
    def display(df): print(df.to_string())

class AnalisisPortafolioHistorico:
    """Análisis cuantitativo con **únicamente** datos históricos reales (sin simulación)."""
    def __init__(self,
                 activos: List[str],
                 benchmark: str = "SPY",
                 fecha_inicio: str = "2023-01-01",
                 fecha_fin: Optional[str] = None,
                 pesos: Optional[Dict[str, float]] = None,
                 tasa_libre_riesgo: float = 0.03,
                 min_observaciones: int = 100,
                 reintentos: int = 3,
                 permitir_simulacion: bool = False):  # forzado a False por defecto
        # entradas
        self.activos = [a.upper().strip() for a in activos if str(a).strip()]
        if not self.activos:
            raise ValueError("Debe proporcionar al menos un activo válido")
        self.benchmark = benchmark.upper().strip()
        self.fecha_inicio = fecha_inicio
        self.fecha_fin = fecha_fin
        self.tasa_libre_riesgo = tasa_libre_riesgo
        self.min_observaciones = int(min_observaciones)
        self.reintentos = int(reintentos)
        self.permitir_simulacion = permitir_simulacion  # NO usar en prod
        # pesos normalizados
        if pesos is None:
            self.pesos = {a: 1.0/len(self.activos) for a in self.activos}
        else:
            s = sum(pesos.values()) or 1.0
            self.pesos = {k: v/s for k, v in pesos.items() if k in self.activos}
        # parámetros
        self.dias_habiles = 252
        self.nivel_confianza = 0.05
        self.z_score_extremo = 3.0
        # datos
        self.precios: Optional[pd.DataFrame] = None
        self.rendimientos: Optional[pd.DataFrame] = None
        self.datos_cargados = False
        self.es_simulado = False  # bandera explícita

    def cargar_datos_historicos(self) -> None:
        """Descarga precios (Close) de yfinance y valida longitud mínima."""
        if self.datos_cargados:
            return
        tickers = self.activos + [self.benchmark]
        ultimo_error = None
        for i in range(self.reintentos):
            try:
                print(f"Descargando datos (intento {i+1}/{self.reintentos})...")
                raw = yf.download(tickers, start=self.fecha_inicio, end=self.fecha_fin,
                                  auto_adjust=True, progress=False)
                if raw is None or raw.empty:
                    ultimo_error = RuntimeError("Descarga vacía")
                    time.sleep(1)
                    continue
                # extrae Close de multiindex o dataframe simple
                precios = raw["Close"] if isinstance(raw.columns, pd.MultiIndex) else raw
                precios = precios.dropna(how='all').copy()
                # eliminar columnas completamente vacías
                precios = precios.dropna(axis=1, how='all')
                # forward-fill de huecos puntuales
                precios = precios.ffill().dropna()
                if len(precios) < self.min_observaciones:
                    ultimo_error = RuntimeError(f"Observaciones insuficientes: {len(precios)} < {self.min_observaciones}")
                    time.sleep(1)
                    continue
                # verifica existencia de todos los tickers solicitados
                faltantes = [t for t in tickers if t not in precios.columns]
                if faltantes:
                    logger.warning(f"Tickers faltantes: {faltantes}")
                self.precios = precios
                # ===== CAMBIO ÚNICO: rendimientos logarítmicos =====
                self.rendimientos = np.log(precios / precios.shift(1)).dropna()
                # ====================================================
                self.datos_cargados = True
                self.es_simulado = False
                print(f"Datos cargados: {len(precios)} obs | columnas: {list(precios.columns)}")
                return
            except Exception as e:
                ultimo_error = e
                logger.error(f"Error de descarga: {e}")
                time.sleep(1)
        # si no hay datos válidos después de reintentos → error duro
        if not self.permitir_simulacion:
            raise RuntimeError(f"No se pudo obtener histórico válido: {ultimo_error}")
        else:
            # rama opcional, deshabilitada por defecto
            print("ADVERTENCIA: modo simulación activado (no recomendado)")
            self._generar_datos_simulados()

    def _generar_datos_simulados(self) -> None:
        """Solo si permitir_simulacion=True (no recomendado para prod)."""
        fechas = pd.date_range(start=self.fecha_inicio, end=self.fecha_fin or "2024-12-31", freq='B')
        np.random.seed(42)
        datos = {}
        for t in self.activos + [self.benchmark]:
            mu, sig = 0.08, 0.20
            n = len(fechas)
            r = np.random.normal(mu/self.dias_habiles, sig/np.sqrt(self.dias_habiles), n)
            p = [100.0]
            for x in r:
                p.append(p[-1]*(1+x))
            datos[t] = p[1:]
        self.precios = pd.DataFrame(datos, index=fechas)
        # ===== CAMBIO ÚNICO: rendimientos logarítmicos =====
        self.rendimientos = np.log(self.precios / self.precios.shift(1)).dropna()
        # ====================================================
        self.datos_cargados = True
        self.es_simulado = True

    def calcular_rendimientos_portafolio(self) -> pd.Series:
        """R_p = sum(w_i * r_i); re-normaliza si faltan columnas."""
        if not self.datos_cargados:
            self.cargar_datos_historicos()
        cols = [a for a in self.activos if a in self.rendimientos.columns]
        w = {a: self.pesos[a] for a in cols}
        s = sum(w.values()) or 1.0
        w = {k: v/s for k, v in w.items()}
        rp = pd.Series(0.0, index=self.rendimientos.index)
        for a, peso in w.items():
            rp = rp.add(peso * self.rendimientos[a], fill_value=0.0)
        return rp.rename("PORTAFOLIO")

    def calcular_metricas_activo(self, r: pd.Series, nombre: str) -> Dict[str, float]:
        """Métricas anualizadas + CAPM (beta/alpha)."""
        if len(r) == 0:
            return {"activo": nombre, "error": "Sin datos"}
        R = r.mean()*self.dias_habiles
        s = r.std()*np.sqrt(self.dias_habiles)
        sharpe = (R - self.tasa_libre_riesgo)/(s + 1e-12)
        acc = (1+r).cumprod(); dd = acc/acc.cummax() - 1
        mdd = float(dd.min())
        dn = r[r<0].std()*np.sqrt(self.dias_habiles)
        sortino = (R - self.tasa_libre_riesgo)/(dn + 1e-12)
        calmar = abs((R - self.tasa_libre_riesgo)/(mdd if mdd != 0 else np.nan))
        beta = np.nan; alpha = np.nan
        if self.benchmark in self.rendimientos.columns and nombre != self.benchmark:
            rb = self.rendimientos[self.benchmark].reindex_like(r).dropna()
            r_al = r.reindex_like(rb).dropna()
            if len(rb) and len(r_al):
                beta = r_al.cov(rb)/(rb.var()+1e-12)
                Rm = rb.mean()*self.dias_habiles
                alpha = R - (self.tasa_libre_riesgo + beta*(Rm - self.tasa_libre_riesgo))
        return {
            "activo": nombre,
            "rendimiento_anual": float(R),
            "volatilidad_anual": float(s),
            "sharpe_ratio": float(sharpe),
            "sortino_ratio": float(sortino),
            "calmar_ratio": float(calmar),
            "max_drawdown": float(mdd),
            "beta": float(beta) if pd.notna(beta) else np.nan,
            "alpha_jensen": float(alpha) if pd.notna(alpha) else np.nan,
            "sesgo": float(r.skew()),
            "curtosis": float(r.kurtosis()),
            "var_5": float(r.quantile(0.05)),
            "observaciones": int(len(r))
        }

    def analisis_estadistico(self) -> None:
        """Descriptivas, Jarque–Bera y Ljung–Box sobre rendimientos."""
        if not self.datos_cargados:
            self.cargar_datos_historicos()
        rp = self.calcular_rendimientos_portafolio()
        series = {a: self.rendimientos[a] for a in self.activos if a in self.rendimientos.columns}
        series["PORTAFOLIO"] = rp
        if self.benchmark in self.rendimientos.columns:
            series[self.benchmark] = self.rendimientos[self.benchmark]
        # descriptivas
        display(pd.DataFrame([{ 'Activo': n,
                                'Media': f"{s.mean():.6f}", 'Mediana': f"{s.median():.6f}", 'Desv_Estándar': f"{s.std():.6f}",
                                'Asimetría': f"{s.skew():.4f}", 'Curtosis': f"{s.kurtosis():.4f}", 'Mínimo': f"{s.min():.6f}",
                                'Máximo': f"{s.max():.6f}", 'VaR_5%': f"{s.quantile(0.05):.6f}", 'Observaciones': len(s)}
                              for n,s in series.items()]))
        # JB
        jb = []
        for n,s in series.items():
            try:
                st,pv = stats.jarque_bera(s.dropna())
                jb.append({'Activo': n, 'Estadístico_JB': f"{st:.4f}", 'P-valor': f"{pv:.6f}", 'Normal_5%': 'Sí' if pv>self.nivel_confianza else 'No'})
            except Exception:
                jb.append({'Activo': n, 'Estadístico_JB': 'Error', 'P-valor': 'Error', 'Normal_5%': 'N/A'})
        display(pd.DataFrame(jb))
        # Ljung–Box
        lb = []
        for n,s in series.items():
            try:
                pv = float(acorr_ljungbox(s.dropna(), lags=10, return_df=True)['lb_pvalue'].iloc[-1])
                lb.append({'Activo': n, 'LB_P-valor': f"{pv:.6f}", 'Autocorrelación_5%': 'Sí' if pv<self.nivel_confianza else 'No'})
            except Exception:
                lb.append({'Activo': n, 'LB_P-valor': 'Error', 'Autocorrelación_5%': 'N/A'})
        display(pd.DataFrame(lb))

    def cambios_y_eventos(self) -> None:
        """CUSUM sobre residuos OLS(r~const) + outliers |z|≥3."""
        if not self.datos_cargados:
            self.cargar_datos_historicos()
        rp = self.calcular_rendimientos_portafolio()
        series = self.rendimientos.copy(); series["PORTAFOLIO"] = rp
        # CUSUM
        cambios = []
        for col in series.columns:
            s = series[col].dropna()
            if len(s) < self.min_observaciones:
                cambios.append({'Activo': col, 'Test': 'CUSUM', 'Estadístico': 'N/A', 'P-valor': 'N/A', 'Resultado': 'Datos insuficientes'})
                continue
            try:
                X = np.ones((len(s),1))
                m = sm.OLS(s.values, X).fit()
                stat,pv,_ = breaks_cusumolsresid(m.resid, ddof=1)
                cambios.append({'Activo': col, 'Test': 'CUSUM', 'Estadístico': f"{stat:.4f}", 'P-valor': f"{pv:.6f}", 'Resultado': 'Estable' if pv>self.nivel_confianza else 'Cambio estructural'})
            except Exception:
                cambios.append({'Activo': col, 'Test': 'CUSUM', 'Estadístico': 'Error', 'P-valor': 'Error', 'Resultado': 'Error'})
        display(pd.DataFrame(cambios))
        # outliers 3σ
        eventos = []
        for col in series.columns:
            s = series[col].dropna()
            if len(s)==0:
                continue
            mu, sd = s.mean(), s.std()+1e-12
            z = abs((s - mu)/sd)
            outs = s[z >= self.z_score_extremo]
            for f,val in outs.items():
                eventos.append({'Activo': col, 'Fecha': f.strftime('%Y-%m-%d'), 'Rendimiento': f"{val:.4f}", 'Rendimiento_%': f"{val*100:.2f}%", 'Z-Score': f"{z.loc[f]:.2f}", 'Tipo': 'Ganancia extrema' if val>0 else 'Pérdida extrema'})
        if eventos:
            df_e = pd.DataFrame(eventos)
            df_e['Z-Score_num'] = pd.to_numeric(df_e['Z-Score'])
            display(df_e.nlargest(20,'Z-Score_num').drop(columns='Z-Score_num'))

    def comparacion_benchmark(self) -> None:
        """Métricas por activo/portafolio + tracking CAPM vs benchmark."""
        if not self.datos_cargados:
            self.cargar_datos_historicos()
        rp = self.calcular_rendimientos_portafolio()
        res = []
        for a in self.activos:
            if a in self.rendimientos.columns:
                res.append(self.calcular_metricas_activo(self.rendimientos[a], a))
        res.append(self.calcular_metricas_activo(rp, "PORTAFOLIO"))
        if self.benchmark in self.rendimientos.columns:
            res.append(self.calcular_metricas_activo(self.rendimientos[self.benchmark], "BENCHMARK"))
        df = pd.DataFrame(res).set_index('activo')
        # formato
        disp = df.copy()
        for c in ['rendimiento_anual','volatilidad_anual','max_drawdown']:
            if c in disp.columns:
                disp[c] = disp[c].apply(lambda x: f"{x:.2%}" if pd.notna(x) else 'N/A')
        for c in ['sharpe_ratio','sortino_ratio','calmar_ratio','beta','alpha_jensen','sesgo','curtosis','var_5']:
            if c in disp.columns:
                disp[c] = disp[c].apply(lambda x: f"{x:.4f}" if pd.notna(x) else 'N/A')
        print("\nMÉTRICAS DE RENDIMIENTO Y RIESGO (históricos reales)")
        display(disp)
        # tracking vs benchmark
        if self.benchmark in self.rendimientos.columns:
            rb = self.rendimientos[self.benchmark]
            excess = rp - rb
            te = excess.std()*np.sqrt(self.dias_habiles)
            corr = rp.corr(rb)
            beta = rp.cov(rb)/(rb.var()+1e-12)
            Rp = rp.mean()*self.dias_habiles
            Rm = rb.mean()*self.dias_habiles
            alpha = Rp - (self.tasa_libre_riesgo + beta*(Rm - self.tasa_libre_riesgo))
            ir = alpha/(te + 1e-12)
            treynor = (Rp - self.tasa_libre_riesgo)/(beta + 1e-12)
            display(pd.DataFrame({
                'Métrica': ['Tracking Error (anual)','Correlación','Beta','Alpha de Jensen (anual)','Information Ratio','Treynor Ratio','Rf usado'],
                'Valor': [f"{te:.2%}", f"{corr:.4f}", f"{beta:.4f}", f"{alpha:.2%}", f"{ir:.4f}", f"{treynor:.4f}", f"{self.tasa_libre_riesgo:.2%}"]
            }))

    def visualizaciones(self) -> None:
        """Rend. acumulados, P vs B, drawdown, distribución."""
        if not self.datos_cargados:
            self.cargar_datos_historicos()
        rp = self.calcular_rendimientos_portafolio()
        # 1) acumulados por activo
        plt.figure(figsize=(8,4))
        for a in self.activos:
            if a in self.rendimientos.columns:
                ra = (1 + self.rendimientos[a]).cumprod() - 1
                plt.plot(ra.index, ra*100, label=a, linewidth=1.2)
        plt.title("Rendimientos acumulados por activo (histórico)")
        plt.xlabel("Fecha"); plt.ylabel("% acumulado")
        plt.legend(bbox_to_anchor=(1.05,1), loc='upper left'); plt.grid(True, alpha=0.3)
        plt.tight_layout(); plt.show()
        # 2) Portafolio vs Benchmark
        plt.figure(figsize=(8,4))
        rp_acc = (1 + rp).cumprod() - 1
        plt.plot(rp_acc.index, rp_acc*100, label='Portafolio', linewidth=2.2)
        if self.benchmark in self.rendimientos.columns:
            rb_acc = (1 + self.rendimientos[self.benchmark]).cumprod() - 1
            plt.plot(rb_acc.index, rb_acc*100, label=f'Benchmark ({self.benchmark})', linewidth=2)
        plt.title("Portafolio vs Benchmark (histórico)")
        plt.xlabel("Fecha"); plt.ylabel("% acumulado")
        plt.legend(); plt.grid(True, alpha=0.3); plt.tight_layout(); plt.show()
        # 3) Drawdown del portafolio
        acc = (1 + rp).cumprod(); dd = acc/acc.cummax() - 1
        plt.figure(figsize=(8,3))
        plt.fill_between(dd.index, dd*100, 0, alpha=0.7, label='Drawdown')
        plt.axhline(y=0, color='black', linewidth=1)
        plt.title("Drawdown del portafolio (histórico)")
        plt.xlabel("Fecha"); plt.ylabel("%")
        plt.legend(); plt.grid(True, alpha=0.3); plt.tight_layout(); plt.show()
        # 4) Distribución + QQ
        plt.figure(figsize=(11,4))
        plt.subplot(1,3,1)
        rp.hist(bins=50, alpha=0.7, edgecolor='black')
        plt.axvline(rp.mean(), color='red', linestyle='--', linewidth=2)
        plt.title("Distribución - Portafolio")
        plt.subplot(1,3,2)
        if self.benchmark in self.rendimientos.columns:
            rb = self.rendimientos[self.benchmark]
            rb.hist(bins=50, alpha=0.7, edgecolor='black')
            plt.axvline(rb.mean(), color='red', linestyle='--', linewidth=2)
            plt.title(f"Distribución - {self.benchmark}")
        plt.subplot(1,3,3)
        stats.probplot(rp, dist="norm", plot=plt)
        plt.title("Q-Q Portafolio vs Normal")
        plt.tight_layout(); plt.show()

    # =======================
    # NUEVO: BOS individual (todos los activos válidos)
    # =======================
    def bos_individual(self, order: int = 2) -> None:
        """BOS por activo y portafolio usando swings en Close (o riqueza Ptf)."""
        if not self.datos_cargados:
            self.cargar_datos_historicos()

        def _bos_from_price(price: pd.Series) -> pd.DataFrame:
            s = price.dropna()
            # swings
            hi_idx = argrelextrema(s.values, np.greater, order=order)[0]
            lo_idx = argrelextrema(s.values, np.less,    order=order)[0]
            swing_high = pd.Series(False, index=s.index)
            swing_low  = pd.Series(False, index=s.index)
            swing_high.iloc[hi_idx] = True
            swing_low.iloc[lo_idx]  = True
            last_sh = s.where(swing_high).ffill()
            last_sl = s.where(swing_low ).ffill()
            ref_sh = last_sh.shift(1); ref_sl = last_sl.shift(1)
            BOS_up   = s > ref_sh
            BOS_down = s < ref_sl
            out = pd.DataFrame({
                "price": s, "swing_high": swing_high, "swing_low": swing_low,
                "last_sh": last_sh, "last_sl": last_sl,
                "BOS_up": BOS_up, "BOS_down": BOS_down
            })
            return out

        # Portafolio: riqueza base 100 desde log-returns
        rp = self.calcular_rendimientos_portafolio()
        wealth_p = np.exp(rp.cumsum()) * 100.0
        datos = {}

        # Activos (sin filtros de longitud)
        for a in self.activos:
            if a in self.precios.columns:
                datos[a] = _bos_from_price(self.precios[a].dropna())
        # Portafolio
        datos["PORTAFOLIO"] = _bos_from_price(wealth_p)

        # Gráficos
        for nombre, dfb in datos.items():
            if dfb.empty:
                continue
            plt.figure(figsize=(8,3))
            plt.plot(dfb.index, dfb["price"], linewidth=1.4, label="Precio/Riqueza")
            plt.scatter(dfb.index[dfb["swing_high"]], dfb.loc[dfb["swing_high"],"price"], s=25, marker="^", label="Swing High")
            plt.scatter(dfb.index[dfb["swing_low" ]], dfb.loc[dfb["swing_low" ],"price"], s=25, marker="v", label="Swing Low")
            up_idx = dfb.index[dfb["BOS_up"]]
            dn_idx = dfb.index[dfb["BOS_down"]]
            plt.scatter(up_idx, dfb.loc[up_idx,"price"], s=20, label="BOS Up")
            plt.scatter(dn_idx, dfb.loc[dn_idx,"price"], s=20, label="BOS Down")
            plt.title(f"BOS – {nombre}")
            plt.xlabel("Fecha"); plt.ylabel("Precio / Riqueza")
            plt.grid(True, alpha=0.3); plt.legend(bbox_to_anchor=(1.02,1), loc="upper left")
            plt.tight_layout(); plt.show()

    # =======================
    # NUEVO: Drawdown individual (desde log-returns)
    # =======================
    def drawdown_individual(self) -> None:
        """Drawdown por activo y portafolio usando riqueza exp(cumsum(log-returns))."""
        if not self.datos_cargados:
            self.cargar_datos_historicos()

        # Portafolio
        rp = self.calcular_rendimientos_portafolio()
        wealth_p = np.exp(rp.cumsum())
        dd_p = wealth_p/wealth_p.cummax() - 1
        plt.figure(figsize=(4,2))
        plt.fill_between(dd_p.index, dd_p*100, 0, alpha=0.7, label="Ptf DD")
        plt.axhline(0, color="k", lw=1)
        plt.title("Drawdown – Portafolio (desde log)"); plt.ylabel("%"); plt.grid(True, alpha=0.3)
        plt.tight_layout(); plt.show()

        # Activos (sin filtros)
        for a in self.activos:
            if a not in self.rendimientos.columns:
                continue
            r = self.rendimientos[a].dropna()
            wealth = np.exp(r.cumsum())
            dd = wealth/wealth.cummax() - 1
            plt.figure(figsize=(4,3))
            plt.fill_between(dd.index, dd*100, 0, alpha=0.6)
            plt.axhline(0, color="k", lw=1)
            plt.title(f"Drawdown – {a} (desde log)"); plt.ylabel("%"); plt.grid(True, alpha=0.3)
            plt.tight_layout(); plt.show()

    # =======================
    # NUEVO: Security Market Line (intercepto fijo en Rf) + regresión anclada
    # =======================
    def security_market_line(self) -> None:
        """
        SML con intercepto fijo en Rf (self.tasa_libre_riesgo) y regresión OLS anclada:
          y = Rf + b*beta  <=>  (y - Rf) = b*beta
        Eje X: β_i; Eje Y: E[R_i]_anual (aritmético) desde media de log-returns.
        """
        if not self.datos_cargados:
            self.cargar_datos_historicos()
        if self.benchmark not in self.rendimientos.columns:
            print("Benchmark no disponible para SML.")
            return

        dias = self.dias_habiles
        Rf = 0.03  # intercepto fijo en el gráfico (3%)
        rb = self.rendimientos[self.benchmark].dropna()
        rp = self.calcular_rendimientos_portafolio().reindex(rb.index).dropna()
        rb = rb.reindex(rp.index).dropna()

        def E_anual_from_log(s: pd.Series) -> float:
            return float(np.expm1(s.mean() * dias))  # aritmético anual desde log

        # puntos por activo (todos válidos)
        betas, rets, labels = [], [], []
        for a in self.activos:
            if a not in self.rendimientos.columns:
                continue
            ri = self.rendimientos[a].reindex(rb.index).dropna()
            if ri.empty:
                continue
            beta_i = float(ri.cov(rb) / (rb.var() + 1e-12))
            ERi = E_anual_from_log(ri)
            betas.append(beta_i); rets.append(ERi); labels.append(a)

        # punto del portafolio
        beta_p = float(rp.cov(rb) / (rb.var() + 1e-12)) if not rp.empty else np.nan
        ERp    = E_anual_from_log(rp) if not rp.empty else np.nan

        # SML teórica
        ERm = E_anual_from_log(rb)
        slope_theory = ERm - Rf

        if len(betas) == 0:
            print("SML: no hay puntos para graficar (revisa datos).")
            return

        betas_arr = np.array(betas, dtype=float)
        y = np.array(rets, dtype=float)

        # OLS anclada: b = sum(beta_i*(y_i - Rf)) / sum(beta_i^2)
        y_mr = y - Rf
        num = float((betas_arr * y_mr).sum())
        den = float((betas_arr**2).sum() + 1e-12)
        slope_ols = num / den

        # Métricas in-sample
        yhat_ols = Rf + slope_ols*betas_arr
        ss_res = float(((y - yhat_ols)**2).sum())
        ss_tot = float(((y - y.mean())**2).sum() + 1e-12)
        r2 = 1 - ss_res/ss_tot

        print("\nSML (intercepto fijo)")
        print(f"  Rf = {Rf:.2%} | ERm = {ERm:.2%} | Pendiente teórica = {slope_theory:.2%}")
        print(f"  OLS anclada -> pendiente={slope_ols:.4f} | R2={r2:.4f}")

        # Plot
        x_min = 0.0
        x_max = max(1.5, (np.nanmax(betas_arr) if len(betas_arr) else 1.0)*1.1)
        xs = np.linspace(x_min, x_max, 200)

        plt.figure(figsize=(6,3))
        plt.plot(xs, Rf + slope_theory*xs, lw=2.0, label=f"SML teórica (Rf={Rf:.0%}, ERm={ERm:.0%})")
        plt.plot(xs, Rf + slope_ols*xs, lw=2.0, linestyle="--", label=f"Regresión anclada (pend={slope_ols:.2f})")

        plt.scatter(betas_arr, y, s=50, label="Activos")
        for b, r, lab in zip(betas_arr, y, labels):
            plt.annotate(lab, (b, r), xytext=(5,5), textcoords="offset points", fontsize=9)

        if np.isfinite(beta_p) and np.isfinite(ERp):
            plt.scatter([beta_p], [ERp], s=80, marker='D', label="Portafolio")

        plt.axhline(Rf, lw=1)
        plt.xlabel("Beta (β)"); plt.ylabel("Rendimiento anual esperado")
        plt.title("Security Market Line (teórica vs regresión anclada)")
        plt.grid(True, alpha=0.3); plt.legend(); plt.tight_layout(); plt.show()

    def ejecutar(self) -> Dict[str, Any]:
        """Pipeline con **solo históricos reales**; error si no hay datos."""
        t0 = time.time()
        print("="*80); print("ANÁLISIS CON HISTÓRICOS REALES (sin simulación)"); print("="*80)
        print(f"Activos: {', '.join(self.activos)} | Benchmark: {self.benchmark}")
        print(f"Rf: {self.tasa_libre_riesgo:.2%} | Período: {self.fecha_inicio} → {self.fecha_fin or 'presente'}")
        self.cargar_datos_historicos()  # puede lanzar RuntimeError
        self.analisis_estadistico()
        self.cambios_y_eventos()
        self.comparacion_benchmark()
        self.visualizaciones()

        # ===== Añadidos solicitados =====
        self.bos_individual(order=2)
        self.drawdown_individual()
        self.security_market_line()

        dt = time.time() - t0
        print("\n"+"="*80); print("COMPLETADO")
        print(f"Tiempo: {dt:.2f}s | Observaciones: {len(self.rendimientos)} | Simulado: {self.es_simulado}")
        print("="*80)
        return {
            'tiempo_ejecucion': dt,
            'observaciones': len(self.rendimientos),
            'activos_analizados': len([a for a in self.activos if a in self.rendimientos.columns]),
            'periodo_analizado': f"{self.rendimientos.index[0].date()} a {self.rendimientos.index[-1].date()}",
            'simulado': self.es_simulado
        }

if __name__ == "__main__":
    activos_opt = ["CSCO","GLD","ORCL","V","XLC"]
    pesos_opt = {"CSCO":0.2694, "GLD":0.1871, "ORCL":0.0268, "V":0.2948, "XLC":0.2219}
    analisis = AnalisisPortafolioHistorico(
        activos=activos_opt,
        benchmark="SPY",
        fecha_inicio="2023-01-01",
        pesos=pesos_opt,
        tasa_libre_riesgo=0.03,
        min_observaciones=100,
        reintentos=3,
        permitir_simulacion=False  # fuerza históricos reales
    )
    resultados = analisis.ejecutar()
```
